```python
import pandas as pd
```

```python
 def cleanData(file):
    df = pd.read_csv(f'{file}.csv')
    df = df.loc[~((df['birth year'] == 1969) & (df['gender']==0))]
    df.to_csv(f'{file}_cleaned.csv', index=False)
    return df
    
cleanData('data/201908-citibike-tripdata')
cleanData('data/201912-citibike-tripdata')
cleanData('data/202008-citibike-tripdata')
cleanData('data/202012-citibike-tripdata')
```
